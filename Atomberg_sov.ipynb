```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'brand':      ['atomberg', 'havells', 'orient', 'crompton', 'usha', 'luminous', 'panasonic'],
    'mentions':   [56, 11, 4, 3, 1, 0, 0],   # total mentions found
    'positive':   [6, 1, 1, 0, 0, 0, 0],      # positive sentiment count
}
df = pd.DataFrame(data)
df['engagement'] = df['mentions'] + df['positive']  # sample formula (modify if needed)
df['M_share'] = df['mentions'] / df['mentions'].sum()
df['E_share'] = df['engagement'] / df['engagement'].sum()
df['SoV_score'] = (df['M_share'] + df['E_share']) / 2   # final Share of Voice Score
print(" Final Table Output:\n")
print(df)
plt.figure(figsize=(8,5))
plt.bar(df['brand'], df['SoV_score'])
plt.title("Share of Voice (SoV) Score")
plt.xlabel("Brand")
plt.ylabel("SOV Score")
plt.show()
plt.figure(figsize=(8,5))
plt.bar(df['brand'], df['mentions'])
plt.title("Brand Mentions (Sample)")
plt.xlabel("Brand")
plt.ylabel("Mentions")
plt.show()
top_brand = df.sort_values(by='SoV_score', ascending=False).iloc[0]
print("\n FINAL INSIGHTS & FINDINGS\n")
print(f" {top_brand['brand']} has the highest Share of Voice dominance with {top_brand['mentions']} mentions.")
print(f" Market comparison shows {top_brand['brand']} leads ahead of competitors in engagement & online presence.")
print(" Recommendation: Focus on content, influencer reach & engagement to increase SoV for lower brands.")
```
